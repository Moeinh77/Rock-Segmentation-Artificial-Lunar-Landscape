**we have to convert the masks to grey scale for using he UNetLearner from fastai**


```python

import time
start=time.time()
import PIL.Image as PilImage

def getClassValues(label_names):

    containedValues = set([])

    for i in range(len(label_names)):
        tmp = open_mask(label_names[i])
        tmp = tmp.data.numpy().flatten()
        tmp = set(tmp)
        containedValues = containedValues.union(tmp)
    
    return list(containedValues)

def replaceMaskValuesFromZeroToN(mask, 
                                 containedValues):

    numberOfClasses = len(containedValues)
    newMask = np.zeros(mask.shape)

    for i in range(numberOfClasses):
        newMask[mask == containedValues[i]] = i
    
    return newMask

def convertMaskToPilAndSave(mask, 
                            saveTo):

    imageSize = mask.squeeze().shape

    im = PilImage.new('L',(imageSize[1],imageSize[0]))
    im.putdata(mask.astype('uint8').ravel())
    im.save(saveTo)

def convertMasksToGrayscaleZeroToN(pathToLabels,
                                   saveToPath):

    label_names = get_image_files(pathToLabels)
    containedValues = getClassValues(label_names)

    for currentFile in label_names:
        currentMask = open_mask(currentFile).data.numpy()
        convertedMask = replaceMaskValuesFromZeroToN(currentMask, containedValues)
        convertMaskToPilAndSave(convertedMask, saveToPath/f'{currentFile.name}')
        print('converted ',currentFile,' !')
    
    print('Conversion finished!')
    
convertMasksToGrayscaleZeroToN(Path('../input/images/clean'), Path('../input/images/clean'))
end=time.time()
print((end-start)/60)
```

```python
# getting the corresponding mask
get_y_fn = lambda x: path_lbl/f'clean{x.stem[6:]}{x.suffix}'
```

**an exmple from the data**

```python
img_f = fnames[2]
img = open_image(img_f)
img.show(figsize=(5,5))
```

```python
mask=open_mask(get_y_fn(img_f))
mask.show(figsize=(5,5))
```

```python
# the max size of bs depends on the available GPU RAM
free = gpu_mem_get_free_no_cache()
if free > 8200: bs=8
else:           bs=4
print(f"using bs={bs}, have {free}MB of GPU RAM free")
```

```python
src_size = np.array(mask.shape[1:])
size=src_size//2
size
```

```python
codes=np.array(['large_rock','smaller_rock','sky','else'],dtype='<U17')
src = (SegmentationItemList.from_folder(path_img)
       .split_by_rand_pct(0.25,seed=1)
       .label_from_func(get_y_fn, classes=codes))
```

```python
data = (src.transform(get_transforms(), size=size, tfm_y=True)# tfm_y does all the transformations on y as well
        .databunch(bs=1)
        .normalize(imagenet_stats))
```

```python
# train
data.show_batch(5, figsize=(7,7))
```

```python
data
```

```python
# defining the accuracy function for the segmentation
def seg_accuracy(input, target):
    target = target.squeeze(1)
    return (input.argmax(dim=1)==target).float().mean()

metrics=[seg_accuracy]
```

```python
wd=1e-2 # weight decay size
# using pretrained resnet
learn = unet_learner(data, models.resnet34, metrics=metrics, wd=wd)
learn.model_dir="/tmp/model/"
learn.summary()
```

```python
lr_find(learn)
learn.recorder.plot(suggestion=True)
```

```python
lr=2.29E-04
```

```python
learn.fit_one_cycle(6, slice(lr), pct_start=0.9)
```

```python
learn.save('stage-1')
```

```python
learn.unfreeze()
```

```python
lrs = slice(lr/400,lr/4)
```

```python
learn.fit_one_cycle(4, lrs, pct_start=0.9)
```
